```python
from PIL import Image, ImageFilter, ImageOps, ImageFile
from rembg import remove
import os
import matplotlib.pyplot as plt
import cv2
import numpy as np
import mediapipe as mp
```

```python
final_images_path = 'images/final'
categories = os.listdir(final_images_path)
# ImageFile.LOAD_TRUNCATED_IMAGES = False
for cat in categories:
    dir_path = os.path.join(final_images_path, cat)

    for img_name in os.listdir(dir_path):
        img = Image.open(os.path.join(final_images_path, cat, img_name))
        img = remove(img)
        img = img.convert('L')
        img.save(os.path.join('images', 'final preprocessed', cat, img_name))
        # plt.imshow(img, cmap='gray')
        # plt.show()
```
